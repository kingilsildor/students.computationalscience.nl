```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/CLS_student_survey_answers.csv")
```

```python
teacher_df = df.filter(like="Which teacher", axis=1)

# Step 1: Split strings by comma if they are strings
teacher_df = teacher_df.applymap(lambda x: x.split(',') if isinstance(x, str) else [])

# Step 2: Count each value per column
value_counts = {
    col: pd.Series([item.strip() for sublist in teacher_df[col] if isinstance(sublist, list) for item in sublist]).value_counts()
    for col in teacher_df.columns
}

# Step 3: Combine into DataFrame
result = pd.DataFrame(value_counts).fillna(0).astype(int)
result
```

```python
questions = result.columns
for question in questions:
    print(result[question].sort_values(ascending=False).head(3))
    print("\n")
```
